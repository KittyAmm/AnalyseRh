```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import cx_Oracle
import sqlalchemy
import scipy
import scipy.stats as stat
import time
%matplotlib inline
from sqlalchemy.exc import SQLAlchemyError
try:
    engine = sqlalchemy.create_engine("oracle+cx_oracle://rhdata:rhdata@localhost?service_name=RH",
                                      arraysize=1000)

    # retrive data from GP_AFFECTATION table
    # affectation_sql = """select NUM_IM,CODEBUR,LIEU_AF,MOTIF_AFFECT,FONCTION_AF,DIVISION_AF,SERVICE_AF,DIRECTION_AF,DATE_PS from GP_AFFECTATION where date_ps IS NOT NULL"""
    affectation_sql = """select NUM_IM,CODEBUR,LIEU_AF,MOTIF_AFFECT,FONCTION_AF,DIVISION_AF,SERVICE_AF,DIRECTION_AF,DATE_PS from GP_AFFECTATION"""

    affectationTest = pd.read_sql(affectation_sql, engine)
    data = affectationTest.copy()
    data.info()
    n = 7
    data.head(int(len(data)*(n/100)))
    dfo = pd.DataFrame(data,  index=data.index, columns=data.columns)
    # détermine si valeur en paramètre est manquante:
    def num_missing(x):
        return sum(x.isnull())
    dfo.apply(num_missing,axis=0)
    dfo.apply(num_missing,axis=1).head()
    # les individus ou lignes
    dfo = dfo.dropna(axis=0)
    # les variables ou colonnes
    dfo = dfo.dropna(axis=1)
    # Supprimer toutes les observations contenant une valeur manquante (au moins sur une variable)
    dfo.dropna(how="all", inplace=True) # lorsqu'on veut garder la même table
    nan_replaced = dfo.replace(np.nan,0)
    # print(nan_replaced)
    dfo['Annee'] = dfo['date_ps'].apply(lambda d: d.year)
    dfo.head()
    xAnne=dfo['Annee'].describe()
    xAnneint = xAnne.astype(int)
    print(xAnneint)
    xService=dfo['service_af'].describe()
    print(xService)


    fig, ax = plt.subplots(figsize=(10,6)) # taille du graphique largeur/hauteur
    plt.plot(dfo['Annee'],dfo['service_af']) # représentation linéaire
    ax.set_ylabel('Services')
    ax.set_xlabel('Date de prise de service')
    ax.set_title('Evolution service') #Titre du graphique
    plt.legend(['Valeur service'],loc='upper left')
    plt.grid()
    plt.show()
    plt.cla()
    plt.clf()

    xservice =pd.value_counts(dfo['service_af'])
    print("tri service: ",xservice)
    xdate=pd.value_counts(dfo['Annee'])
    print("tri date: ",xdate)

      #Regardons maintenant la service tendance en calculant le pourcentage ou le taux.
    x = dfo['service_af'].mode().value_counts()
    # print("mode :", x)
    dico=dict(x) # Transformer x en un dictionnaire
    # print("dico :",dico)
    # Transformation du dictionnaire en dataframe
    col = dico.keys() # récupérer les clés du dictionnaires (Ces clés représentent les modalités analysées. Elles seront les variables dans le dataframe à créer).
    print("col :",col)
    df = pd.DataFrame(dico,columns=col, index=[0]) # index=[0] doit être spécifié lorsque les valeurs du dictionnaire ne sont pas encadré par [].
    df['total']=df.sum(axis=1)
    for names, values in df.iteritems(): # on récupère le nom de la colonne et sa valeur
        df['{name}'.format(name=names)]=100*df['{name}'.format(name=names)]/df['total'] # Calcul du pourcentage
        # print("total :",df['total'])
        # print("name :",df['{name}'.format(name=names)])
    # print(df) # on obtient ainsi le tableau de fréquence en pourcentage
    ax = df.plot(kind='barh')
    plt.title("Tendance de service")
    plt.show()

    # Durée de travail dans une service
    services = dfo.groupby(by=['service_af'], as_index=False).count()
    # sb.set_style("white")
    # # servicebyDate = services.groupby(by='date_ps').count()
    # p = sb.catplot(x = "date_ps", y = "service_af", hue = "lieu_af", data = services, kind = "strip",palette = "Set2", height= 4, aspect = 2)
    # p.fig.suptitle("Année par service selon les dates de prise de poste")
    # p.set_axis_labels("Année", "Services")
    # plt.show()

    # tri croisé sur colonne service_af et date_ps afin d'obtenir un tableau de contingence(fréquences absolues)
    Anneint = dfo['Annee'].astype(int)
    serviceDate = pd.crosstab(Anneint,['service_af']).fillna(0).copy()
    print("service selon date :",serviceDate)
    # Fréquences relatives
    # Pourcentage par rapport à la ligne
    xsum = pd.crosstab(Anneint,['service_af']).apply(lambda r: r/r.sum(),axis=1)
    # Pourcentage par rapport à la colonne
    xsumc = pd.crosstab(Anneint,['service_af']).apply(lambda r: r/r.sum(),axis=0)
    # Pourcentage par rapport au total
    xtotal = serviceDate.apply(lambda r: r/len(df),axis=0)
    # print(xtotal)
    sb.set(style="darkgrid")
    sb.histplot(data=serviceDate, x=Anneint, color="green", label="Frequence dans service", kde=True)
    plt.legend()
    plt.show()
    sb.set_style("whitegrid")
    total = sb.boxplot(x = dfo['service_af'], y = 'Annee', data = dfo)
    print("total :", total)
    plt.show()

    st_chi2, st_p, st_dof, st_exp = stat.chi2_contingency(xtotal)
    print("chi_2: ",st_chi2)
    print("p_value: ",st_p)
    print("degre liberte: ",st_dof)
    print("tab freq: ",st_exp)

    sb.histplot(data=serviceDate, x="Annee", hue="service_af")
    plt.title("Contingence dans service")
    plt.show()


    # Mesure d’association entre variables qualitatives : le test d'indépendance de khi-deux (stat khi-deux, pvalue, ddl, et tableau de fréquence théorique)
    # print("chi2 :",stat.chi2_contingency(serviceDate))
    X = dfo['Annee']
    print("Annee", X)
    Y = dfo['service_af']
    cont = dfo[['Annee', 'service_af']].pivot_table(index=X, columns=Y, aggfunc=len).fillna(0).copy()
    tx = X.value_counts()
    ty = Y.value_counts()
    cont = cont.astype(int)
    # print(cont)
    # st_chi2, st_p, st_dof, st_exp = stat.chi2_contingency(serviceDate)
    st_chi2, st_p, st_dof, st_exp = stat.chi2_contingency(cont)
    print("chi_2: ",st_chi2)
    print("p_value: ",st_p)
    print("degre liberte: ",st_dof)
    print("tab freq: ",st_exp)


    # fonctions de répartition
    median_val=[serviceDate['service_af'].median()]
    print("Mean : ",stat.percentileofscore(serviceDate['service_af'],median_val,kind='mean'))
    # Le test exact de Fisher
    # print(stat.fisher_exact(serviceDate)) # (20.0, 0.034965034965034919) oddsratio et pvalue

except SQLAlchemyError as e:
    print(e)
```
