```python
try:
    engine = sqlalchemy.create_engine("oracle+cx_oracle://rhdata:rhdata@localhost?service_name=RH",
                                      arraysize=1000)
    # retrive data from GP_AFFECTATION et GP_AFFECTATION
    bureauAffect_sql = """SELECT DIRECTION_AF,ACTIF_BUR FROM GP_BUREAU b join GP_AFFECTATION a on b.codebur=a.codebur"""
    bureauAffect = pd.read_sql(bureauAffect_sql, engine)
    # dimension
    bureauAffect.shape
    # exploration
    bureauAffect.head()
    data_group = bureauAffect.groupby(by=['direction_af'], as_index=False).count()
    ax = sb.barplot(x="direction_af", y="actif_bur", data=data_group)
    plt.xticks(rotation=90)
except SQLAlchemyError as e:
    print(e)
```
